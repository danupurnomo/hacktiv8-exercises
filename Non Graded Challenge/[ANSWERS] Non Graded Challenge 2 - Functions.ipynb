**Summary**
```
Kelvin to Celsius: C = K - 273 (C = K - 273.15 if you want to be more precise)

Kelvin to Fahrenheit: F = 9/5(K - 273) + 32 or F = 1.8(K - 273) + 32

Celsius to Fahrenheit: F = 9/5(C) + 32 or F = 1.80(C) + 32

Celsius to Kelvin: K = C + 273 (or K = C + 271.15 to be more precise)

Fahrenheit to Celsius: C = (F - 32)/1.80

Fahrenheit to Kelvin: K = 5/9(F - 32) + 273.15
```

### Problem 1

**Buatlah sebuah function yang dapat mengkonversi suhu dari kelvin ke celcius, dan celcius ke kelvin.**

Formula to convert `Kelvin` to `Celcius`.

> °C = °K - 273.15

````python
# Function to Convert Kelvin to Celcius

def kelvin_to_celcius(temp):
  '''
  This function is used to convert Kelvin to Celcius.
  
  :param `temp` : Kelvin's temperatue | int or float.
  :return: `temp` : Celcius's temperature | float.
  '''

  temp = temp - 273.15

  return temp
````

Formula to convert `Celcius` to `Kevin`.

> °K = °C + 273.15

````python
# Function to Convert Celcius to Kelvin

def celcius_to_kelvin(temp):
  '''
  This function is used to convert Celcius to Kelvin.
  :param `temp` : Celcius's temperatue | int or float.
  :return: `temp` : Kelvin's temperature | float.
  '''
  
  temp = temp + 273.15
  
  return temp
````

````python
# Trying Functions

temp_kelvin = 373.15
temp_celcius = 100

print('Kelvin to Celcius')
print('Kelvin temperature value  : ', temp_kelvin)
print('Celcius temperature value : ', kelvin_to_celcius(temp_kelvin))
print('')

print('Celcius to Kelvin')
print('Celcius temperature value : ', temp_celcius)
print('Kelvin temperature value  : ', celcius_to_kelvin(temp_celcius))
````

### Problem 2

**Buatlah sebuah function yang dapat mengkonversi suhu ke fahrenheit. Tambahkan parameter untuk memastikan bahwa argumen yang dimasukan adalah celcius atau kelvin. Panggil function yang pertama jika diperlukan.**

Formula to convert `Celcius` to `Fahrenheit`.

> °F = 9/5 * (°C) + 32

````python
# Function to Convert Celcius to Fahrenheit

def celcius_to_fahrenheit(temp):
  '''
  This function is used to convert Celcius to Fahrenheit.
  
  :param `temp` : Celcius's temperatue | int or float.
  :return: `temp` : Fahrenheit's temperature | float.
  '''

  temp = (9/5*temp) + 32

  return temp
````

````python
# Function to Convert Temperature to Fahrenheit

def convert_to_fahrenheit(temp, origin='celcius'):
  '''
  This function is used to convert Celcius/Kelvin to Fahrenheit.
  
  :param `temp` : Celcius's Temperature or Kelvin's Temperature | int or float
  :param `origin` : Type of origin temperature (celcius or kelvin)| string
  :return: `temp` : Fahrenheit's Temperature | float
  '''

  if origin == 'celcius':
    temp = celcius_to_fahrenheit(temp)
  elif origin == 'kelvin':
    temp = kelvin_to_celcius(temp)
    temp = celcius_to_fahrenheit(temp)
  else: 
    temp = 'None'
    print('Unknown value of param `origin`. Param `origin` must be `celcius` or `kelvin`')
  
  return temp
````

````python
# Trying Functions

temp = 100
origin_1 = 'celcius'
origin_2 = 'kelvin'
origin_3 = 'reamur'

print('Celcius to Fahrenheit')
print('Celcius temperature value    : ', temp)
print('Fahrenheit temperature value : ', convert_to_fahrenheit(temp=temp, origin=origin_1))
print('')

print('Kelvin to Fahrenheit')
print('Kelvin temperature value     : ', temp)
print('Fahrenheit temperature value : ', convert_to_fahrenheit(temp=temp, origin=origin_2))
print('')

print('Reamur to Fahrenheit')
print('Reamur temperature value     : ', temp)
print('Fahrenheit temperature value : ', convert_to_fahrenheit(temp=temp, origin=origin_3))
````

### Problem 3

**Buatlah sebuah function yang dapat mengkonversi suhu dari fahrenheit. Berikan argumen untuk memastikan bahwa outputnya adalah celcius atau kelvin.**

Formula to convert `Fahrenheit` to `Celcius`.

> °C = (°F - 32) / (9/5)

````python
# Function to Convert Fahrenheit to Celcius

def fahrenheit_to_celcius(temp):
  '''
  This function is used to convert Fahrenheit to Celcius.
  
  :param `temp` : Fahrenheit's temperatue | int or float.
  :return: `temp` : Celcius's temperature | float.
  '''

  temp = (temp - 32) / (9/5)

  return temp
````

````python
# Function to Convert Temperature from Fahrenheit

def convert_from_fahrenheit(temp, dest='celcius'):
  '''
  This function is used to convert Fahrenheit to Celcius/Kelvin.
  
  :param `temp` : Fahrenheit's Temperature | int or float
  :param `dest` : Type of destination temperature (celcius or kelvin)| string
  :return: `temp` : Celcius's/Kelvin's Temperature | float
  '''

  if dest == 'celcius':
    temp = fahrenheit_to_celcius(temp)
  elif dest == 'kelvin':
    temp = fahrenheit_to_celcius(temp)
    temp = celcius_to_kelvin(temp)
  else: 
    temp = 'None'
    print('Unknown value of param `dest`. Param `dest` must be `celcius` or `kelvin`')
  
  return temp
````

````python
# Trying Functions

temp = 212
dest_1 = 'celcius'
dest_2 = 'kelvin'
dest_3 = 'reamur'

print('Fahrenheit to Celcius')
print('Fahrenheit temperature value : ', temp)
print('Celcius temperature value    : ', convert_from_fahrenheit(temp=temp, dest=dest_1))
print('')

print('Fahrenheit to Kelvin')
print('Fahrenheit temperature value : ', temp)
print('Kelvin temperature value     : ', convert_from_fahrenheit(temp=temp, dest=dest_2))
print('')

print('Fahrenheit to Reamur')
print('Fahrenheit temperature value : ', temp)
print('Reamur temperature value     : ', convert_from_fahrenheit(temp=temp, dest=dest_3))
````

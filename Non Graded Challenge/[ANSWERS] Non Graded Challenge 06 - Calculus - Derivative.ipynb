## Assignment Instructions

- Carilah derivative dari sebuah fungsi:
  - $y = x^{2} + 2x + 1$
  - $y = 4x^{3} - 3x^{2} + 2x -1$

---
## Answer

```python
# Define the Equations

import sympy as sy

x = sy.Symbol('x')

equation_1 = x**2 + 2*x + 1
equation_2 = 4*x**3 - 3*x**2 + 2*x -1
```

```python
# Display First Equation

equation_1
```

```python
# Display First Derivative of First Equation

equation_1.diff()
```

```python
# Display Second Equation

equation_2
```

```python
# Display First Derivative of Second Equation

equation_2.diff()
```

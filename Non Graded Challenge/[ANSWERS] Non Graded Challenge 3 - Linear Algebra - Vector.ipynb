### Problem 2

2. Diberikan 3 buah vektor 3 dimensi:
$$
A=
\begin{bmatrix}
17 \\
22 \\
19
\end{bmatrix}
,B=
\begin{bmatrix}
10 \\
20 \\
11
\end{bmatrix},
C=
\begin{bmatrix}
5 \\
12 \\
9
\end{bmatrix}
$$

Hitunglah:
  - A+B
  - B-C
  - A dot C
  - A x B
  - norm A
  - Sudut antara vektor A dan B

```python
# Define Vectors

A = np.array([17, 22, 19])
B = np.array([10, 20, 11])
C = np.array([5, 12, 9])

print('Vector A : ', A, type(A))
print('Vector B : ', B, type(B))
print('Vector C : ', C, type(C))
```

```python
# A + B

A + B
```

```python
# B - C

B - C
```

```python
# A dot C

A @ C
```

```python
# Norm A

np.linalg.norm(A)
```

```python
# Angle between A and B

## First, let's find Cosine Similarity from Dot Product
result_dot_product = A @ B
cos_theta = result_dot_product / (np.linalg.norm(A) * np.linalg.norm(B))

# Change `cos_theta` into radian
radians = np.arccos(cos_theta)

# Change radians into angle
degree = np.degrees(radians)

print('Cos Theta    : ', cos_theta)
print('Radians      : ', radians)
print('Angle/Degree : ', degree)
```

### Problem 3

3. Buatlah plot dari vector berikut ini kedalam bidang 2D:

  $$
  U =
  \begin{bmatrix}
  2 \\ 5
  \end{bmatrix},
  V =
  \begin{bmatrix}
  3 \\ 1
  \end{bmatrix}
  $$

```python
# Define the Vectors

U = np.array([2, 5])
V = np.array([3, 1])

print('Vector U : ', U, type(U))
print('Vector V : ', V, type(V))
```

```python
# Plot 2D

import matplotlib.pyplot as plt

plt.arrow(0, 0, U[0], U[1], color="r", head_width=0.2)
plt.arrow(0, 0, V[0], V[1], color="b", head_width=0.2)
plt.axis([0, 7, 0, 7])
plt.grid()
plt.show()
```

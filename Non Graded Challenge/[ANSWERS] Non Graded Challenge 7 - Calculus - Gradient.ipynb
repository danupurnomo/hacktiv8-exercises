## Assignment Instructions

Carilah gradient dari sebuah fungsi:
  - $3x+4y=5$

---
## Answer

### Gradient of `x`

From $3x+4y=5$ we can change it into 

> <center>$y=5/4 - 3/4x$
> <br>
> $y=1.25 - 0.75x$</center>

```python
# Find the Gradient

import numpy as np

## Define `x`
x = np.linspace(-50, 50)

## Define the function
y = 1.25 - 0.75 * x

result_gradient_x = np.gradient(y)
result_gradient_x
```

```python
# Plot `x` and `result_gradient_x`

import matplotlib.pyplot as plt

fig, ax = plt.subplots(ncols=2, figsize=(21,5))

ax[0].plot(x, y)
ax[0].set_title('Plot of `x`')

ax[1].plot(result_gradient_x, y)
ax[1].set_title('Plot of `result_gradient_x`')
plt.xlim(-2, 2)
```

---
### Gradient of `y`

From $3x+4y=5$ we can change it into $x=5/3 - 4/3y$

```python
# Find the Gradient

## Define `y`
y = np.linspace(-50, 50)

## Define the function
x = (5/3) - (4/3) * x

result_gradient_y = np.gradient(x)
result_gradient_y
```

```python
# Plot `y` and `result_gradient_y`

import matplotlib.pyplot as plt

fig, ax = plt.subplots(ncols=2, figsize=(21,5))

ax[0].plot(x, y)
ax[0].set_title('Plot of `y`')

ax[1].plot(x, result_gradient_y)
ax[1].set_title('Plot of `result_gradient_y`')
plt.ylim(-3, 3)
```

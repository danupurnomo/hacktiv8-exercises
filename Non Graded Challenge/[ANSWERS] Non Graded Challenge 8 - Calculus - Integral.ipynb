## Assignment Instructions

Carilah integral dari sebuah fungsi:
  - $\int \! (3x^{2}-6x+3) \, \mathrm{d}x.$
  - $\int \! (8x^{3}-x^{2}+5x-1) \, \mathrm{d}x.$

---
## Answer

```python
# Find the Integral

import sympy as sy

x = sy.Symbol('x')
equation_1 = 3*x**2 - 6*x + 3
equation_2 = 8*x**3 - x**2 + 5*x -1

integral_equation_1 = sy.integrate(equation_1)
integral_equation_2 = sy.integrate(equation_2)
```

```python
# Display `integral_equation_1`

integral_equation_1
```

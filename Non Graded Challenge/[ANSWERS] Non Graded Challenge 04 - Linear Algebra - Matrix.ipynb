### Problem 3

Lakukan Transpose, hitung determinan, dan inverse terhadap matrix dibawah ini:

  $$
  \begin{bmatrix}
  23 & 50 & 19 \\
  7 & 12 & 109 \\
  57 & 67 & 98
  \end{bmatrix}
  $$

```python
# Define the Matrix

vect_3 = np.array([ 
                   [23, 50, 19],
                   [7, 12, 109],
                   [57, 67, 98]
                   ])
```

```python
# Transpose

vect_3.T
```

```python
# Determinant

np.linalg.det(vect_3)
```

```python
# Inverse

np.linalg.inv(vect_3)
```
